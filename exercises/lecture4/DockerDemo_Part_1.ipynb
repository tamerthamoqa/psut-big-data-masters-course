##Create Notebook

> Create notebook and name it ```find_pi.ipynb``` to estimate $\pi = 3.14$ via Simulation.


<center><img src="https://www.101computing.net/wp/wp-content/uploads/estimating-pi-monte-carlo-method.png" alt="PSUT"  width="180px">
</center>

```bash 
π ≈ 4 x (number of points in the circle / total number of points)
```

```python 
# Spark Pi Example 
import sys
from random import random
from operator import add
from pyspark.sql import SparkSession

spark = SparkSession\
    .builder\
    .appName("PythonPi")\
    .getOrCreate()

n_points = 100000   # Total number of simulation points 
n_points_array=range(1, n_points + 1)

def f(_):
    x = random() * 2 - 1   # x will be in range [-1,1]
    y = random() * 2 - 1   # y will be in range [-1,1]
    within_circle=1 if x ** 2 + y ** 2 <= 1 else 0  # check if the point is located within unit circle 
    return within_circle

# Calculate number of points in a circle using map-reduce
n_points_in_circle = spark.sparkContext.parallelize(n_points_array).map(f).reduce(add)

print("Pi is roughly %f" % (4.0 * n_points_in_circle / n_points))
```
